We create the set of target vectors from the training set T = {{0,0,1,...},{0,1,0...}}

```python
import pandas as pd
cp = pd.read_csv('coupon_detail_train.csv')
print(cp)
```

```python
T = []
for i in range(int(round(0.3*cp["COUPON_ID_hash"].size))):
  index = users.USER_ID_hash[users.USER_ID_hash == cp.loc[i,"USER_ID_hash"]].index.tolist()
  if index != []:
    t = dictionary_coupons[cp.loc[i,"COUPON_ID_hash"]]

    T.append(t)

print(T)
```

We design a neural network with the appropriate characteristics for the training set.

Whose output is a softwax multiple (returns a vector of probabilities)

```python
from keras import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU
from keras.layers import PReLU
from keras.layers import ELU
from keras.layers import Dropout
from keras.utils.vis_utils import plot_model

def build_regressor():
  regressor = Sequential()
  regressor.add(Dense(units=2, input_shape=(2,)))
  regressor.add(ELU(alpha=1))
  regressor.add(Dense(units=int(round(sizes)/3)))
  regressor.add(ELU(alpha=1))
  regressor.add(Dropout(0.25))
  regressor.add(Dense(units= sizes,activation='softmax'))
  regressor.compile(optimizer='Nadam', loss='categorical_crossentropy',metrics=['accuracy'])
  return regressor

build_regressor().summary()


iterations = 20
from keras.wrappers.scikit_learn import KerasRegressor
regressor = KerasRegressor(build_fn=build_regressor,batch_size=20,epochs=iterations,validation_split=0.2,verbose=2)

import time
tic = time.time()
results = regressor.fit(np.array(x).np.array(T))
print("Processing Time (secs):", time.time()-tic)
```

```python
import random
import numpy as np


# new users
age = random.randint(1,100)
sex = np.random.choice(2)
if sex == 0:
  sex_str = 'M'
else:
  sex_str = 'F'


# predict the probabilities of buying (coupon_dictionary)
# number of existing coupons

arr = regressor.predict(np.array([age,sex]))
sum = 0
for i in range(1,11):
  sum += np.partition(arr.flatten(), -2)[-i]
if sum > 0.1:
  gurantee = 'YES'
else:
  gurantee = 'NO'

# find the most likely coupon to buy

max_prob = np.amax(arr)
result = np.where(arr==np.amax(arr))
sizes = df["COUPON_ID_hash"].size
vector = np.zeros(sizes)
vector[result[1][0]] = 1

#print('Max prob',max_prob)
#print('Returned tuple of arrays:', result)

print("AGE:",age)
print("SEX:", sex_str)
print("Coupon most probable to be sold:", [k for k,v in dictionary_coupons.items() if np.array_equal(v,vector)])
print("Prob of selling the coupon most probable to be sold", round(max_prob*100,3),"%")
print("Prob of selling any of the 10 coupons most probable to be sold", round(sum*100),"%")
print("Almost Ensured sell?", gurantee)
```
